```python
## TARGET : Diabete or not ?
targets = df.loc[:, "Outcome"].unique()
numberTargerts = df.loc[:, "Outcome"].value_counts().values



## Plot
fig, ax = plt.subplots()

labels = ["Yes", "No"]

ax.pie(numberTargerts, labels = labels, autopct = '%1.1f%%')
ax.set_title('Diabete ?')


plt.tight_layout()
plt.show()
```
